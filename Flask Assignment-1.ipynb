Q1. What is Flask Framework? What are the advantages of Flask Framework?

Flask is a lightweight and flexible web framework for Python. It is designed to be simple and easy to use,
providing the essentials needed for web development without imposing too many constraints. Flask is classified as
a micro-framework because it doesn't include built-in tools and libraries for tasks like database abstraction and
form validation, but it allows developers to choose their preferred tools for these purposes.

Advantages :-

Simplicity: Flask is known for its simplicity and ease of use. It has a straightforward API, making it accessible to both beginners and experienced developers.

Flexibility: Flask is unopinionated and doesn't force a particular structure on developers.

Extensibility: Flask can be easily extended using Flask extensions or by integrating with third-party libraries. 

Community and Documentation: Flask has a strong and active community, and there is extensive documentation available.

Lightweight: Being a micro-framework, Flask is lightweight and doesn't come with unnecessary features.

Easy to Learn: Flask's simplicity makes it easy to learn for developers who are new to web development or Python.

Compatibility with Python Libraries: Flask seamlessly integrates with other Python libraries, allowing developers to leverage the broader Python ecosystem for various tasks.

Routing: Flask provides a simple and intuitive way to define URL patterns and associate them with specific functions (views).
    
In Flask, app routing refers to the process of defining URL patterns and associating them with specific view functions in a 
Flask application. Routing determines how the application responds to different HTTP requests (such as GET or POST) for 
specific URLs. Flask uses the "@app.route" decorator to define routes.

 app routing in Flask is a fundamental aspect of building web applications. It provides a way to define how different URLs 
should be handled, making the application modular, organized, and easy to maintain.

Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the
url_for() function.

```python
In Flask, the url_for() function is used for URL building. It generates a URL for a given endpoint by dynamically
creating a URL based on the provided endpoint name and any arguments or keyword arguments associated with the endpoint.

from flask import Flask, url_for, redirect

app = Flask(__name__)

@app.route('/home')
def home():
    return 'Welcome to the Home Page'

@app.route('/about')
def about():
    return 'This is the About Page'

@app.route('/visit')
def visit():
    home_url = url_for('home')
    about_url = url_for('about')

    return f'To visit the Home Page, go to: {home_url}<br>To visit the About Page, go to: {about_url}'

if __name__ == "__main__":
    app.run(debug=True)
    
    
# The /home and /about routes define two simple endpoints (home and about).
# The /visit route uses the url_for() function to dynamically generate URLs for the home and about endpoints.
# The generated URLs are then displayed when you access the /visit route.
```
