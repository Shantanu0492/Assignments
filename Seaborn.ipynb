The five plots in seaborn are as below:
    
sns.relpot :   
This plot give the relation between to datas present in the dataset. It's designed to visualize the relationship between 
two numerical variables. It is a figure-level function, meaning it creates a new figure each time it's called.

sns.scatterplot :
This plot is use to visualize the numerical data on the plot in scatter way. By using this plot we can analysis the data 
that how much it is scatter accros the plot or in a straight line. By using the scatter plot we can decide the data pattern.

sns.pairplot:
Plots pairwise relationships in a dataset, showing scatter plots for numerical variables and histograms for univariate
distributions.

sns.jointplot :
Jointplot gives the two plot at a time, it gives the histogram + scatter plot pattern of the data.

sns.catplot :
sns.catplot function in Seaborn is a figure-level function for drawing categorical plots onto a FaceGrid.
