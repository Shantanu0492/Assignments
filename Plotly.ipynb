```python
Q1. Load the "titanic" dataset using the load_dataset function of seaborn. Use Plotly express to plot a
scatter plot for age and fare columns in the titanic dataset.
```

```python
import pandas as pd
import plotly.graph_objects as go
import matplotlib.pyplot as plt
```

```python
titanic = pd.read_csv(r"D:\Data\titanic.csv")
titanic.head()
```

```python
# Use Plotly express to plot a scatter plot for age and fare columns in the titanic dataset.

fig = go.Figure()
fig.add_trace(go.Scatter(x=titanic.age, y=titanic.fare, mode='markers'))
fig.show()
```

```python
tips = pd.read_csv(r"D:\Data\tips.csv")
tips.head()
```

```python
import seaborn as sns
import plotly.express as px

# Create a box plot with Plotly Express
fig = px.box(tips, x='day', y='total_bill', points="all", title='Box Plot of Total Bill by Day')
fig.show()
```

```python
Q3. Using the tips dataset in the Plotly library, Plot a histogram for x= "sex" and y="total_bill" column in
the tips dataset. Also, use the "smoker" column with the pattern_shape parameter and the "day"
column with the color parameter.
```

```python
fig = go.Figure()
fig = px.histogram(
    tips,
    x='sex',
    y='total_bill',
    color='day',
    facet_col='smoker',
    pattern_shape='smoker',
    labels={'total_bill': 'Total Bill'},
    title='Histogram of Total Bill by Sex, Day, and Smoker Status'
)
fig.show()
```

```python
Q4. Using the iris dataset in the Plotly library, Plot a scatter matrix plot, using the "species" column for
the color parameter.
Note: Use "sepal_length", "sepal_width", "petal_length", "petal_width" columns only with the
dimensions parameter.
```

```python
iris = px.data.iris()
fig = px.scatter_matrix(iris,dimensions=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'],color='species',
    title='Scatter Matrix Plot of Iris Dataset',
    labels={'sepal_length': 'Sepal Length', 'sepal_width': 'Sepal Width', 'petal_length': 'Petal Length', 'petal_width': 'Petal Width'}
)
fig.show()
```

```python
Q5. What is Distplot? Using Plotly express, plot a distplot.
```

```python
A distplot (distribution plot) is a type of plot that combines a histogram with a kernel density estimate (KDE) plot, 
providing a visual representation of the distribution of a univariate dataset. It shows the frequency distribution of 
continuous data and estimates the probability density function.
```

```python
import numpy as np
np.random.seed(42)
data = np.random.normal(size=1000)
fig = px.histogram(data, marginal='rug', title='Distplot Example', labels={'value': 'Data Points'})
fig.show()
```
