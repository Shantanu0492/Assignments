Q4. What is ‘DataFrame’ in pandas and how is it different from pandas.series? Explain with an example.

A DataFrame is a two-dimensional, tabular data structure that consists of columns and rows. It is similar to a
spreadsheet or SQL table, where data is organized in a grid-like format. Each column in a DataFrame can be of a different 
data type, and you can performvarious operations on both rows and columns.

On the other hand, a Series is a one-dimensional array-like object that can hold any data type. It is essentially a single 
column of a DataFrame, along with an index. A DataFrame, therefore, can be thought of as a collection of Series, where each 
Series represents a column.


Example : -

```python
#Series
series_data = pd.Series([10, 20, 30, 40], name='Column_A')
print("Series:")
print(series_data)
print("\n")

#DataFrame
df_data = pd.DataFrame({
    'Column_A': [10, 20, 30, 40],
    'Column_B': ['A', 'B', 'C', 'D']
})
print("DataFrame:")
print(df_data)
```

Q6. Which of the following is mutable in nature Series, DataFrame, Panel?

Series: Series is mutable, meaning you can modify its values, index, or other properties after it has been created.

DataFrame: DataFrame is also mutable. You can add, remove, or modify columns and rows in a DataFrame after it has been 
created.

Panel: Panel was a three-dimensional data structure in Pandas for handling three-dimensional data. However, as of version
0.25.0, the Panel class has been deprecated, and developers are encouraged to use MultiIndex DataFrames or the xarray library
for three-dimensional data.

```python
# Creating multiple series

series1 = pd.Series([1,2,3,4,5])
series2 = pd.Series(["Shantanu","Rajat","Ayush","Rahul","Rajesh"])
series3 = pd.Series(["Pune","Nagpur","Pune","Shimla","Bhopal"])

dataframe_1 = pd.DataFrame({"Sr.No":series1,
                           "Student_Name":series2,
                           "Location":series3})
dataframe_1
```

In this example, I created three Series (series1, series2, and series3). Each Series represents a column, and they are
combined into a DataFrame using a dictionary.
The resulting DataFrame has three columns: 'Sr.No', 'Student_Name', and 'Location'. Each column contains the data from 
the respective Series.
