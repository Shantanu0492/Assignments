Q1. Explain GET and POST methods.

Both GET and POST are HTTP methods used to request and send data between a client and a server. 
They are fundamental to web development and are commonly used in the context of HTML forms.

The GET method is used to request data from a specified resource. Parameters are appended to the URL as key-value pairs.
Parameters are visible in the URL, and they are limited by the length restrictions of URLs.

The POST method is used to submit data to be processed to a specified resource. Parameters are sent in the request body, 
not in the URL. They are not visible in the URL. Parameters are not visible in the URL, providing a level of privacy.

In Flask, the request object is used to access incoming request data when handling HTTP requests. 
It provides a convenient way for developers to interact with data sent by clients (such as browsers) to the server. 
The request object allows you to retrieve form data, query parameters, cookies, and other information associated with
the current HTTP request.

Q3. Why is redirect() used in Flask?

In Flask, the redirect() function is used to perform a redirect to a different URL. It is a utility function that 
simplifies the process of redirecting a user from one route to another. The redirect() function is particularly 
useful when you want to send the user to a different endpoint after they perform a certain action or when you want
to implement a URL redirection.

The redirect() function simplifies the process of triggering a client-side redirect in response to a request. It helps in keeping the code clean and enhances code readability.

Q4. What are templates in Flask? Why is the render_template() function used?

In Flask, templates are used to separate the presentation (HTML, CSS, etc.) from the application logic. 
Templates allow you to create dynamic web pages by embedding variables and control structures within HTML, 
making it easier to manage and update the user interface.

The render_template() function in Flask is used to render templates by combining them with the data provided. 
It takes the name of the template file as its first argument and any additional arguments representing data that 
should be passed to the template.

```python
from flask import Flask,request ,render_template , jsonify

app = Flask(__name__)


@app.route('/')
def home_page():
    return render_template('index.html')

@app.route('/postman_action',methods=['POST'])
def math_ops1():
    if(request.method == 'POST'):
        ops = request.json['operation']
        num1 = int(request.json['num1'])
        num2 = int(request.json['num2'])
        if ops == 'add':
            r = num1+num2
            result = "The sum of " + str(num1) + 'and ' + str(num2) + "is " + str(r)
        if ops == 'subtract':
            r = num1-num2
            result = "The subtract of " + str(num1) + 'and ' + str(num2) + "is " + str(r)
        if ops == 'multiply':
            r = num1*num2
            result = "The multiply of " + str(num1) + 'and ' + str(num2) + "is " + str(r)
        if ops == 'divide':
            r = num1/num2
            result = "The divide of " + str(num1) + 'and ' + str(num2) + "is " + str(r)
            
        return jsonify(result)

if __name__=="__main__":
    app.run(host="0.0.0.0")
```
