```python
Q2. What are glyphs in Bokeh, and how can you add them to a Bokeh plot? Explain with an example.
```

```python
In Bokeh, glyphs are visual representations of data on a plot. They define how the data points should be displayed,
such as using circles, squares, lines, or other shapes. Glyphs play a central role in creating visualizations, allowing 
you to represent data in a meaningful and visually appealing way.

To add glyphs to a Bokeh plot, you typically use the figure object and call glyph-specific methods such as circle(), line(),
or square(). These methods create instances of the corresponding glyph and add them to the plot.
```

```python
from bokeh.plotting import figure, show

# Sample data
x = [1, 2, 3, 4, 5]
y_circle = [6, 7, 2, 4, 5]
y_line = [5, 4, 3, 7, 6]

# Create a Bokeh figure
p = figure(title="Glyphs Example", x_axis_label="X-axis", y_axis_label="Y-axis")

# Add circles to the plot
p.circle(x, y_circle, size=10, color="blue", legend_label="Circles")

# Add a line to the plot
p.line(x, y_line, line_width=2, color="red", legend_label="Line")

# Add a legend
p.legend.location = "top_left"

# Show the plot
show(p)
```

```python
Q3. How can you customize the appearance of a Bokeh plot, including the axes, title, and legend?
```

We can customize the appearance of a Bokeh plot by modifying various attributes of the figure object, 
which represents the overall plot. Here's how you can customize different aspects of a Bokeh plot, including axes, 
title, and legend:

```python
from bokeh.plotting import figure, show

# Sample data
x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]

# Create a Bokeh figure
p = figure(
    title="Customized Bokeh Plot",
    x_axis_label="X-axis Label",
    y_axis_label="Y-axis Label",
    width=600,
    height=400,
    toolbar_location='above',  # Move the toolbar above the plot
)

# Customize the appearance of circles
p.circle(x, y, size=10, color="green", legend_label="Data Points")

# Customize axes
p.xaxis.axis_label_text_color = "blue"
p.yaxis.axis_label_text_font_style = "italic"

# Customize title
p.title.text_color = "purple"
p.title.text_font_size = "20px"

# Customize legend
p.legend.location = "top_left"
p.legend.label_text_font_style = "bold"

# Show the plot
show(p)
```
